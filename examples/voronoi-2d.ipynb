```python
import geogram as geo
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.tri as tri

np.random.seed(0)
```

```python
(vertices, triangles) = geo.shape.quad()

seed_nb = 1000
seeds   = np.random.rand(seed_nb, 2)

voronoi = geo.Voronoi(seeds, domain_vertices=vertices, domain_simplices=triangles)
triang  = tri.Triangulation(voronoi.q[:, 0], voronoi.q[:, 1], voronoi.t)

# Plot
fig, ax = plt.subplots()
ax.set_aspect('equal')

ax.tripcolor(triang, voronoi.tseed, shading='flat')
ax.set_title('Voronoi diagram')
```

```python
weights     = np.array([1. for i in range(seed_nb)])
weights[50] = 1.04

voronoi = geo.Voronoi(seeds, weights, domain_vertices=vertices, domain_simplices=triangles)
triang  = tri.Triangulation(voronoi.q[:, 0], voronoi.q[:, 1], voronoi.t)

# Plot
fig, ax = plt.subplots()
ax.set_aspect('equal')

ax.tripcolor(triang, voronoi.tseed, shading='flat')
ax.set_title('Voronoi diagram')
```
