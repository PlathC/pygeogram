```python
import geogram as geo
import numpy as np
import polyscope as ps

ps.init()
np.random.seed(0)
```

```python
SEED_NB = 1000
```

```python
# Load and compute a volume domain
(vertices, triangles) = geo.shape.cube()
(vertices, tets)      = geo.mesh.tetrahedralize(vertices, triangles)
```

```python
# Com
seeds   = np.random.rand(SEED_NB, 3)
voronoi = geo.Voronoi(seeds, domain_vertices=vertices, domain_simplices=tets)
```

```python
weights      = np.zeros(SEED_NB)
weights[600] = .08

laguerre = geo.Voronoi(seeds, weights, domain_vertices=vertices, domain_simplices=tets)
```

```python
# Display both meshes
ps_vol = ps.register_volume_mesh("Voronoi diagram", voronoi.q, voronoi.t)
ps_vol.add_scalar_quantity("Cell indices", voronoi.tseed, enabled=True, defined_on='cells')
ps_vol.set_cull_whole_elements(True)

ps_vol = ps.register_volume_mesh("Laguerre diagram", laguerre.q, laguerre.t)
ps_vol.add_scalar_quantity("Cell indices", laguerre.tseed, enabled=True, defined_on='cells')
ps_vol.set_position((1.1, 0, 0))
```
