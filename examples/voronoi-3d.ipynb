```python
import geogram as geo
import numpy as np
import polyscope as ps

ps.init()
np.random.seed(0)
```

```python
(vertices, triangles) = geo.shape.cube()
(vertices, tets)      = geo.mesh.tetrahedralize(vertices, triangles)

seed_nb = 1000
seeds   = np.random.rand(seed_nb, 3)
voronoi = geo.Voronoi(seeds, domain_vertices=vertices, domain_simplices=tets)

vertices = voronoi.q[voronoi.t].reshape(-1, 3)
tets = np.array([i for i in range(len(vertices))]).reshape((-1, 4))
indices = [voronoi.tseed[int(i/4)] for i in range(len(vertices))]

ps_vol = ps.register_volume_mesh("Voronoi diagram", vertices, tets)
ps_vol.add_scalar_quantity("Cell indices", indices, enabled=True)
ps_vol.set_cull_whole_elements(True)
```

```python
weights      = np.zeros(seed_nb)
weights[600] = .08

voronoi = geo.Voronoi(seeds, weights, domain_vertices=vertices, domain_simplices=tets)
```

```python
vertices = voronoi.q[voronoi.t].reshape(-1, 3)
tets     = np.arange(0, len(vertices)).reshape((-1, 4))
indices  = voronoi.tseed[(np.arange(0, len(vertices)) / 4).astype(int)]

ps_vol = ps.register_volume_mesh("Laguerre diagram", vertices, tets)
ps_vol.add_scalar_quantity("Cell indices", indices, enabled=True)
ps_vol.set_position((1.1, 0, 0))
```
